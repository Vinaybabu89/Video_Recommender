```python
#to find number of days after publishing
def time(x):
    values=pd.to_datetime(x,format='%Y-%m-%dT%H:%M:%SZ')
    date=values.date()
    #print(type(date))
    
    today=date.today()
    today.strftime('%Y-%m-%d')
    
    now=today-date
    a=str(now)
    a=a.split(',')[0]
    
    return a
    
```

```python
def hours(x):
    values=pd.to_datetime(x,format='%Y-%m-%dT%H:%M:%SZ')
    time=values.time()
    x=str(time)
    hour=datetime.datetime.strptime(x, '%H:%M:%S').strftime('%I:%M %p')
    return hour
```

```python
import re
import nltk
from nltk.corpus import stopwords
#from nltk.corpus import words
from nltk.stem import WordNetLemmatizer 
```

```python
lema=WordNetLemmatizer()
title=[]
for i in range(len(data['Title'])):
    tit=re.sub('[^a-zA-Z]',' ',data['Title'][i])
    tit=tit.lower()
    tit = tit.split()
    tit=[lema.lemmatize(word) for word in tit if not word in set(stopwords.words('english'))]
    tit=" ".join(tit)
    title.append(tit)
    
```

```python
data['Clean_Title']=title
data['Clean_Title']=data['Clean_Title'].fillna('')
```

```python
indices=pd.Series(data.index,index=data['Title']).drop_duplicates()
indices
```

```python
indices==300
```

```python
def rec_tit(title,sig=sig):
    idx=indices[title]
    
    sig_scores=list(enumerate(sig[idx]))
    #sortbased on sigmoid values
    sig_scores=sorted(sig_scores,key=lambda x :x[1],reverse=True)
    
    sig_scores=sig_scores[1:11]
    
    title_indices=[i[0] for i in sig_scores]
    
    Video_tit=data['Title'].iloc[title_indices]
    score=data['Custom_score'].iloc[title_indices]
    Url=data['Video_URL'].iloc[title_indices]
    sig_score=sig_scores
    # input values into dataframe
    rec_df=pd.DataFrame({'VIdeo':Video_tit,'Scores':score,'Url':Url,'S.score':sig_score})
    #resultant dataframe is again sorts based on custom_score
    rec_df=rec_df.sort_values("Scores", ascending=False)
    return rec_df
```
