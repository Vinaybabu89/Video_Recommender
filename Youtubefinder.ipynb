```python
title=[]
channel_id=[]
channel_title=[]
descripition=[]
tags=[]
pusblished_time=[]
views=[]
likes=[]
dislike=[]
tumbnails=[]
comment=[]
URL=[]
subscriber=[]
```

```python
def find_videos(search:str):
    item=[]
    res_VID=[]
    res_SAT=[]
    ch_ID=[]
    res_chid=[]
    
    res = youtube.search().list(
        q=search,part='snippet',type='video',maxResults=50).execute()
    
    for tumb in res['items']:
        tumbnail=tumb['snippet']['thumbnails']['high']['url']
        tumbnails.append(tumbnail)
    
    
    for x in res['items']:
        video_id=x['id']['videoId']
        item.append(video_id)
        
    for ele in item:
        res_vId=youtube.videos().list(id=ele,part='snippet').execute()
        res_VID+=res_vId['items']
        
    for ele_1 in item:
        res_stat=youtube.videos().list(id=ele_1,part='statistics').execute()
        res_SAT+=res_stat['items']
        
    for c_id in res['items']:
        ID=c_id['snippet']['channelId']
        ch_ID.append(ID)
    
    for ids in ch_ID:
            res_cid=youtube.channels().list(id=ids,part='statistics').execute()
            res_chid+=res_cid['items']
                
    for data_cst in res_chid:
        try:
            subc=int(data_cst['statistics']['subscriberCount'])
            subscriber.append(subc)
        except KeyError as c:
            c=None
            subscriber.append(c)
             
    for data in res_VID:
        tit=data['snippet']['title']
        title.append(tit)
        ch_id=data['snippet']['channelId']
        channel_id.append(ch_id)
        ch_tit=data['snippet']['channelTitle']
        channel_title.append(ch_tit)
        des=data['snippet']['localized']['description']
        descripition.append(des)
        pub_T=data['snippet']['publishedAt']
        pusblished_time.append(pub_T)
        
    for data1 in res_VID:
        tag=None
        try:
            tag=data1['snippet']['tags']
        except KeyError as e:
            e=None
        tags.append(tag)
        
        
    for data_st in res_SAT:
        like=None
        d_like=None
        com=None
        try:
            view=data_st['statistics']['viewCount']
            like=data_st['statistics']['likeCount']
            d_like=data_st['statistics']['dislikeCount']
            #fav=data_st['statistics']['favoriteCount']
            #favourite.append(fav)
            com=data_st['statistics']['commentCount'] 
        except KeyError as a:
            a=None
        views.append(view)
        likes.append(like)
        dislike.append(d_like)
        comment.append(com)
                
    
    for video_id in item:
        url="https://www.youtube.com/watch?v=" + video_id
        URL.append(url)
        
        
        
        
        
    dataframe = pd.DataFrame({'Title':title,'Descripition':descripition,'ChannelID':channel_id,'Channel_Title':channel_title,
                              'Tumbnails':tumbnails,'Subscriber_Count':subscriber,'Published_Time':pusblished_time,
                              'Views':views,'Likes':likes,'Dislike':dislike,'Comments': comment,'Video_URL':URL,'TAGS':tags})
    youtube_data=dataframe.to_csv('Grand slam.csv')# this will land csv file where jupyter is installed
    
    return youtube_data
```
